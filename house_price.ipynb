```python
"""
we will be using a data set that compares house size and house prices of properties,
sold to predict the cost of houses based on size.

"""
# Import libraries and use to alias 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load data from CSV file
data = pd.read_csv('C:\\Users\\Administrador\\workspace\\Python\\Machine_learning\\Data\\home_dataset.csv')
 
# Extract features and target variable
house_sizes = data ['HouseSize'].values
house_prices = data['HousePrice'].values
```

```python
# Visualize the data
plt.scatter(house_sizes, house_prices, marker ='o', color ='blue')
plt.title('House Prices vs. House Size')
plt.xlabel('House Size (sq.ft)')
plt.ylabel('House Price ($)')
plt.show()
```

```python
"""
A train-test splits the generated data into training and testing sets
using the train_test_split() function from Sklearn.

The train test split is a model validation procedure commonly used in predictive machine learning
to simulate how a model will work with new/unknown data.
It is commonly used with large data sets or when you need a quick estimate.
"""
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(house_sizes, house_prices, test_size=0.2, random_state=42)

# Reshape the data for NumPy
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Predict prices for the test set
predictions = model.predict(x_test)

# Visualize the predictions
plt.scatter(x_test, y_test, marker='o', color='blue', label='Prices')
plt.plot(x_test, predictions, color='green', linewidth=2, label='Predicted Prices')
plt.title('Price Prediction with Linear Regression')
plt.xlabel('House Size (sq.ft)')
plt.ylabel('House Price (millions $)')
plt.legend()
plt.show()
```
